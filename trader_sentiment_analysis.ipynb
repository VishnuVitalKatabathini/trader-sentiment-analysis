```python
#performing the daily summary
```

```python
daily_summary=trader_sentiment_df.groupby(['date','classification']).agg({'Closed PnL':'sum','Account':'count','Size Tokens':'mean','Size USD':'mean'}).reset_index()
```

```python
daily_summary
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=daily_summary, x='classification', y='Account', estimator='sum')
plt.title('Total Number of Trades by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Total Trades')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=daily_summary, x='classification', y='Closed PnL', estimator='mean',)
plt.title('Average PnL by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Closed PnL')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=daily_summary, x='classification', y='Size Tokens', estimator='mean',)
plt.title('Average Trade Size by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Trade Size')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=daily_summary, x='classification', y='Size USD', estimator='mean',)
plt.title('Average Trade Size by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Trade Size')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=daily_summary.sort_values('date'), x='date', y='Closed PnL', hue='classification')
plt.title('Daily Closed PnL Over Time by Sentiment')
plt.xlabel('Date')
plt.ylabel('Closed PnL')

plt.show()
```

```python
# Average PnL by Sentiment
avg_pnl = trader_sentiment_df.groupby('classification')['Closed PnL'].mean()

avg_pnl.plot(kind='bar', title='Average Closed PnL by Market Sentiment', ylabel='Avg Closed PnL', xlabel='Sentiment', color=['red', 'green', 'gray'])
plt.grid(True)
plt.show()
```

```python
- On average, traders showed higher profitability during "Greed" days compared to "Fear".
- "Neutral" days had more stable but lower average PnL.
- The standard deviation of PnL was highest in "Greed", indicating both big wins and big losses.


- "Greed" periods had a higher number and proportion of profitable trades.
- "Fear" periods had more conservative trades (lower size), often break-even or small losses.



```
